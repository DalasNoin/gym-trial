```python
import gym
import numpy as np
weights = np.random.normal(0.0,0.1,4)

memory = []
seed=10
learning_rate = 0.05
```

```python
def sigmoid(x):
    return 1/(1+np.exp(x))

def length(a):
    x = 0
    for element in a:
        x = x + element
    return x / len(a)

def execute(observation, weights):
    return int(round(sigmoid(length(observation*weights))))

def sort(memory):
    mean = np.sum(np.transpose(np.array(memory))[1]) / len(memory)
    i=0
    while i < len(memory):
        
        if memory[i][1] < mean:
            memory.pop(i)
            
        i += 1
    
    return memory
            

```

```python
memory = []
env = gym.make('CartPole-v0')
for idx in range(5):
    for i_episode in range(seed):
        observation = env.reset()
        if idx == 0:
            weights = np.random.normal(0,0.2,4)
        else:
            weights = memory[i_episode][0] + np.random.normal(0,learning_rate,4)
        for t in range(100):
            env.render()
            print(observation)
            action = execute(observation,weights)
            print(action)
            observation, reward, done, info = env.step(action)
            if(done):
                print("Episode finished after {} timesteps".format(t+1))
                memory.append((weights, t))
                break
    sort(memory)
    seed = len(memory)
```
